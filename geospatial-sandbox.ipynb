```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Grasslands data
grass_url = "https://data.fs.usda.gov/geodata/edw/edw_resources/shp/S_USA.NationalGrassland.zip"

grass_gdf = gpd.read_file("https://data.fs.usda.gov/geodata/edw/edw_resources/shp/S_USA.NationalGrassland.zip")
grass_gdf
```

```python
# States data
us_url = "https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_state_20m.zip"

us_gdf = gpd.read_file(us_url)
us_gdf
```

```python
# AIANNH land areas url
aiannh_url = "https://www2.census.gov/geo/tiger/TIGER2020/AIANNH/tl_2020_us_aiannh.zip"

# Open land area boundaries
aiannh_boundary = gpd.read_file(aiannh_url)
aiannh_boundary
```

```python

# Select SD state boundary
sd = us_gdf[us_gdf["NAME"] == "South Dakota"]
sd

# Clip AIANNH boundary to SD boundary
aiannh_sd = aiannh_boundary.clip(sd.geometry)
aiannh_sd

# Clip Grasslands boundary to SD boundary
grass_sd = grass_gdf.clip(sd.geometry)
grass_sd
```

```python
# Plot the first shapefile with legend
ax = aiannh_sd.plot(column='NAME', legend=True, label = "AIANNH Land Areas")
# Plot the second shapefile with legend on the same axis
grass_sd.plot(column='GRASSLANDN', legend=True, ax=ax, label = "National Grassland Units")
#Plot 3rd shapefile (SD Boundary)
#sd.plot(ax=ax, color='None', edgecolor='grey')
# Set the title and show the plot
plt.title("Two Shapefiles with Legends")
plt.legend()
plt.show()
```
